```python
import pandas as pd


from matplotlib import pyplot as plt
from matplotlib import rcParams
import seaborn as sns
```

```python
#grafico linha
x = range(1,10)
y = [1,2,3,4,4,5,2,3,6,]

plt.plot(y,x)
```
