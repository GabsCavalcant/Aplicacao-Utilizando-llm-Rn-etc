```python
import pandas as pd
import numpy as np

```

```python
df = pd.DataFrame(np.arange(36).reshape(6,6))
```

```python
df
```

```python
df.drop([1,5],axis= 1 )
```

```python
serie = pd.Series(np.arange(6))
serie.name = 'Nova variavel'

serie
```

```python
df = df.join(serie)
```

```python
df.sort_values(by='Nova variavel', ascending=False)
```
