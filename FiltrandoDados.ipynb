```python
import pandas as pd
import numpy as np

from pandas import DataFrame, Series

```

```python
np.random.seed(25)

indice = ['l2','l3','l4','l5','l6']
coluna = ['c2','c3','c4','c5','c6']

df = DataFrame(np.random.rand(25).reshape((5,5)),
               index = indice,
               columns = coluna)
```
