```python
import pandas as pd
import numpy as np

from pandas import DataFrame
```

```python
df = pd.DataFrame(np.arange(36).reshape(6,6))
df
```

```python
df2 = pd.DataFrame(np.arange(15).reshape(5,3))
df2
```

```python
pd.concat([df,df2])
```

```python
pd.concat([df,df2], axis=1)
```
